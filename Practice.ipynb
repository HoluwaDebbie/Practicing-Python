```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
filepath = r'/content/bigmart_data.csv'
```

```python
df = pd.read_csv(filepath)
```

```python
df.head(20)
```

```python
df.T
```

```python
df.columns
```

```python
df.columns = df.columns.str.lower()
```

```python
df.columns
```

```python
df.info()
```

##Problem
### Dropping missing values

```python
df.dropna(subset=['item_weight', 'outlet_size'], inplace=True)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop_duplicates()
```

```python
df.info()
```

```python
df["item_type"].unique()
```

###Checking the relationship between item type and outlet location


```python
pd.crosstab(df['outlet_location_type'], df['item_type'])
```

```python
count_data = df.groupby(["outlet_location_type", "item_type"]).size().unstack()
```

```python
count_data
```

```python
count_data.T
```

```python
#plotting using matplotlib

count_data.plot(kind='bar',figsize=(16,6))
plt.title('Count of Item Types by Outlet Location')
plt.xlabel('Outlet Location')
plt.ylabel('Count')
plt.show()
```

```python
#plotting usind seaborn

plt.figure(figsize=(16,6))
sns.countplot(data=df, x='outlet_location_type', hue='item_type')
plt.title('Count of Item Types by Outlet Location')
plt.xlabel('Outlet Location')
plt.ylabel('Count')
plt.show()
```

###Using same data. Instead of dropping missing values, will be filling missing values with mean and mode.

```python
file_path = r'/content/bigmart_data.csv'
df1 = pd.read_csv(file_path)
```

```python
df1.head()
```

```python
df1.info()
```

###Problems
filling missing values with mean and mode

```python
df1['Item_Weight']= df1['Item_Weight'].fillna(df1['Item_Weight'].mean())
```

```python
df1['Item_Weight']
```

```python
df1['Outlet_Size']= df1['Outlet_Size'].fillna(df1['Outlet_Size'].mode()[0])
```

```python
df1['Outlet_Size']
```

```python
df1.info()
```

```python
df1.describe()
```

```python
pd.crosstab(df1['Outlet_Location_Type'], df1['Item_Type'])
```

```python
count_data1 = pd.crosstab(df1['Outlet_Location_Type'], df1['Item_Type'])
```

```python
count_data1
```

```python
count_data1.T
```

```python
#plotting using matplotlib

count_data1.plot(kind='bar',figsize=(10,6))
plt.title('Count of Item Types by Outlet Location')
plt.xlabel('Outlet Location')
plt.ylabel('Count')
plt.show()
```

```python
#plotting using seaborn

plt.figure(figsize=(10,6))
sns.countplot(data=df1, x='Outlet_Location_Type', hue='Item_Type')
plt.title('Count of Item Types by Outlet Location')
plt.xlabel('Outlet Location')
plt.ylabel('Count')
plt.show()
```
